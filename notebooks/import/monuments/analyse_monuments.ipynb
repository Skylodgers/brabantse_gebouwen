```python
import os
import json
import xmltodict
import pandas as pd


source_file = './source/mapping_full.xml'

# Open and read the XML file
with open(source_file, 'r') as xml_file:
    xml_data = xml_file.read()
    source_dict = xmltodict.parse(xml_data, encoding='utf-8', process_namespaces=False, namespace_separator=':')    
    records = source_dict['records']['record']
with open('./out/monuments.json', 'w') as f:
    json.dump(records, f, indent=2)
    
    print(json.dumps(records[1]['rdf:RDF']['schema:LandmarksOrHistoricalBuildings'], indent=2))
    
```

```python
unique_keys = set()

# Iterate over each dictionary in the list
for dictionary in records:
    # Add the keys of the current dictionary to the set
    unique_keys.update(dictionary['rdf:RDF']['schema:LandmarksOrHistoricalBuildings'].keys())

# Convert the set to a list (if needed)
all_keys_list = list(unique_keys)

print(json.dumps(all_keys_list, indent=2))
```
