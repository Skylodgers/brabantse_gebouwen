```python
import os
from lxml import etree as ET
import json
import xmltodict
import pandas as pd


mapping_def_file = '../source/mapping_edm.xml'

# Open and read the XML file
with open(mapping_def_file, 'r') as xml_file:
    xml_data = xml_file.read()

```

```python
parsed_data = xmltodict.parse(xml_data)
mappings = parsed_data['rec-mapping']['node-mappings']['node-mapping']
mapping_dict = {}
mapping_dict['dataset'] = ['museum-rotterdam__2023_01_30_16_02.sip.zip']
for mapping in mappings:
    mapping_dict[mapping['@outputPath']] = [mapping['@inputPath']]

#print(json.dumps(mapping_dict, indent=2))
mapping_df = pd.DataFrame.from_dict(mapping_dict)
mapping_df
```
